```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

# get_html 테스트

```python
def get_html(url):
    _html = ""
    resp = requests.get(url)
    if resp.status_code == 200:
        _html = resp.text
    return _html

URL = "http://finance.naver.com/sise/sise_index_day.nhn?code=KPI200&page=1"
html = get_html(URL)
soup = BeautifulSoup(html, 'html.parser')
soup
```

# 날짜랑 주가지수 받아오기

```python
code=soup('td',{'class':'number_1'})
index_list = []
date_list = []

for page in range(10):
    URL = "http://finance.naver.com/sise/sise_index_day.nhn?code=KPI200&page={:d}".format(page)
    html = get_html(URL)
    soup=BeautifulSoup(html, 'html.parser')
    soup

    date = soup('td',{'class':'date'})
    for date_text in date:
        date_list.append(str(date_text).split('date">')[1].split("</td")[0])
    code = soup('td',{'class':'number_1'})

    for i, code_text in enumerate(code):
        try:
            if i%4 == 0:
                index_list.append(str(code_text).split('number_1">')[1].split('</td>')[0])
        except:
            print(code_text)
            continue
```

# 좀 더 간단하게 원하는 기업 데이터 받아오기(한개 기업만)
여러 기업에 대해 함께 자료를 받아오고 싶으면 code_list를 고쳐주고 dataframe을 저장하는 방식을 좀 수정해주면 됨. 아래에 완성된 코드가 있음

```python
PAGE_NUM = 10
df_example = pd.DataFrame(columns=['Date', 'Close', 'Open', 'High', 'Low', 'Volume'])


code_list2 = ['005930']

for code in code_list2:
    for num in range(PAGE_NUM):
        # Get Data From URL
        URL = "http://finance.naver.com/item/sise_day.nhn?code={code}&page={page}".format(code=code, page=num+1)
        html = get_html(URL)
        soup = BeautifulSoup(html, 'html.parser')
        meta_date = soup('span',{'class':'tah p10 gray03'})

        # Create meta which contains all data except date and price change
        meta = soup('td', {'class': 'num'})
        meta = [meta[i] for i in range(len(meta)) if i%6 != 1]



        # Parse Data into list
        len_per_page = len(meta_date)
        _df = pd.DataFrame(np.zeros((len_per_page, 6)), columns=['Date', 'Close', 'Open', 'High', 'Low', 'Volume'])
        _date = np.array([str(x).split('gray03">')[1].split('</span')[0].replace(',','') for x in meta_date])
        _price = np.array([str(x).split('p11">')[1].split('</span')[0].replace(',','.') for x in meta])


        # Make a dataframe which will be merged into the main dataframe
        _df.iloc[:len_per_page,0] = _date
        _df.iloc[:len_per_page,1:] = _price.reshape(len_per_page, 5)

        # Append created dataframe into main dataframe
        df_example = df_example.append(_df)

df_example = df_example.iloc[::-1].set_index('Date')
print(df_example)
```

```python
####### KOSPI200 종목 크롤링
target_n = 240

adj_code_list = []
adj_name_list = []
KOSPI200_close = pd.DataFrame()
KOSPI200_volume = pd.DataFrame()

for i in range(len(code_list)):
    # BOOLEAN VALUE which will be used if there are no data available for that stock.
    # If there are no more data available, END_LOOP = TRUE and loop for this stock will stop.
    END_LOOP = False

    # Dataframe which will contain data for code_list[i]
    data_temp = pd.DataFrame(columns=['Date','Open','High','Low','Close','Volume'])

    # Loop : Get a dataframe for a company which has a code 'code_list[i]'
    try:
        for j in range(int(target_n/10)):
            # Get Data From URL
            URL = "http://finance.naver.com/item/sise_day.nhn?code={code}&page={page}".format(code=code_list[i], page=j+1)
            html = get_html(URL)
            soup = BeautifulSoup(html, 'html.parser')
            meta_date = soup('span',{'class':'tah p10 gray03'})

            # Create meta which contains all data except date and price change
            meta = soup('td', {'class': 'num'})

            # if i % 6 == 1, that data is price change
            meta = [meta[i] for i in range(len(meta)) if i%6 != 1]

            # if len_per_page != 10, This means that there are no more data available for that stock.
            # In that case, escape from this loop and continue to loop for next stock.
            # Also, remove empty datas from meta list.
            len_per_page = len(meta_date)
            if len_per_page != 10:
                END_LOOP = True
                meta = meta[: 5 * len_per_page]

            # Parse Data into array
            _df = pd.DataFrame(np.zeros((len_per_page, 6)), columns=['Date', 'Close', 'Open', 'High', 'Low', 'Volume'])
            _date = np.array([str(x).split('gray03">')[1].split('</span')[0].replace(',','') for x in meta_date])
            _price = np.array([str(x).split('p11">')[1].split('</span')[0].replace(',','') for x in meta])

            # Make a dataframe which will be merged into the main dataframe
            _df.iloc[:len_per_page,0] = _date
            _df.iloc[:len_per_page,1:] = _price.reshape(len_per_page, 5)

            # Append created dataframe into main dataframe(data_temp)
            data_temp = data_temp.append(_df)

            if END_LOOP == True:
                break

        # Export Dataframe for code 'code_list[i]'
        data_temp = data_temp.iloc[::-1].set_index("Date")
        KOSPI200_close[code_list[i]] = data_temp.iloc[:,3]
        KOSPI200_volume['data/',code_list[i]] = data_temp.iloc[:,4]
        data_temp.to_csv('data/',code_list[i] +'.csv')

        # Make a list for code and name which are used in crawling. This is needed because we should know
        # Which code&name are used in crawling not causing any errors.
        adj_code_list.append(code_list[i])
        adj_name_list.append(name_list[i])

        print("Loop for {code} is done! '{cnt}/200".format(code=code_list[i], cnt = i+1))

    except Exception as e:
        print("There is an error in {code}".format(code=code_list[i]))
        print("Error: ", e)
        continue

df_port_info = pd.DataFrame({'Code':adj_code_list, 'Name':adj_name_list})
df_port_info.to_csv('data/codenamelist.csv', encoding='cp949')

KOSPI200_close.to_csv('data/KOSPI200_close.csv')
KOSPI200_volume.to_csv('data/KOSPI200_volume.csv')
```
